# Imports

```python
import pandas as pd
import numpy as np
```

# Dataset

```python
df = pd.read_csv("../data/raw/household_power_consumption.txt", sep=";")
```

```python
df.keys()
```

```python
df_standar = df.replace("?", np.nan)
cols = df_standar.columns[2:8]

for c in cols:
    df_standar[c] = pd.to_numeric(df_standar[c], errors="coerce")
```

```python
df_standar["datetime"] = pd.to_datetime(df_standar["Date"] + " " + df_standar["Time"], format="%d/%m/%Y %H:%M:%S")
```
